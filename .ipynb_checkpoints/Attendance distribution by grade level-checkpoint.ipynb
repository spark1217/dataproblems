```python
query = '''SELECT joined.grade_level, COUNT(attendance_status) AS absent_tardy, total_num_students
            FROM joined
            JOIN grades ON joined.grade_level = grades.grade_level
            WHERE attendance_status="Tardy" OR attendance_status="Absent"
            GROUP BY joined.grade_level'''
result = mysql(query)
result
```
