# Categorizing foods
You are given the following dataframe and are asked to categorize each food into 1 of 3 categories: meat, fruit, or other.
Given this, write code to add a new column categorizing each row.

```python
import pandas as pd

data = {'food': ['bacon','STRAWBERRIES','Bacon','STRAWBERRIES','BACON','strawberries','strawberries','pecans'], 'pounds' : [4.0,3.5,7.0,3.0,6.0,9.0,1.0,3.0]}
df = pd.DataFrame(data, columns=['food','pounds'])
```

```python
df
```

```python
category = []
for i in df['food']:
    if i.upper() == 'BACON':
        category.append('meat')
    elif i.upper() == 'STRAWBERRIES':
        category.append('fruit')
    else:
        category.append('other')
df['category'] = category
df
```
